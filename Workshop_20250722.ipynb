ã“ã®ãƒ‡ãƒ¼ã‚¿ã‚’ä½¿ã£ã¦ã€**æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•**ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†ï¼

ã“ã¡ã‚‰ã¯æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ã‚’ä½œæˆã™ã‚‹ã‚³ãƒ¼ãƒ‰ã§ã™ã€‚è©¦ã—ã«å®Ÿè¡Œã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚

```
#####################################
# ğŸ“Š ãƒ‡ãƒ¼ã‚¿ã‚’å…¥åŠ›ã—ã‚ˆã†

days = ['Mon.', 'Tue.', 'Wed.', 'Thu.', 'Fri.']

study_hours = [5, 8, 6, 9, 4]
```

daysã¨ã„ã†éƒ¨åˆ†ã«æ›œæ—¥ã‚’ã€study_hoursã¨ã„ã†éƒ¨åˆ†ã«å‹‰å¼·æ™‚é–“ã‚’ãã‚Œãã‚Œå…¥åŠ›ã§ãã¾ã™ã€‚ã‚«ãƒ³ãƒã§ãã‚Œãã‚ŒåŒºåˆ‡ã‚‰ã‚Œã¦ã„ã¦ã€å·¦ã‹ã‚‰é †ç•ªã«1,2,3,...ã¨ã„ã†ç•ªå·ï¼ˆã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹ï¼‰ãŒä¸ãˆã‚‰ã‚Œã¦ã„ã¾ã™ã€‚ãã®ãŸã‚ï¼Œä¸€ç•ªç›®ãŒæœˆæ›œæ—¥ã®ãƒ‡ãƒ¼ã‚¿ã‚’ï¼ŒäºŒç•ªç›®ãŒç«æ›œæ—¥ã®ãƒ‡ãƒ¼ã‚¿ã‚’ãã‚Œãã‚Œè¡¨ã—ã¾ã™ã€‚

## Step1: èª²é¡Œ

ã“ã¡ã‚‰ã¯ã€æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ã‚’è¦‹ã‚„ã™ãåŠ å·¥ã—ãŸã‚³ãƒ¼ãƒ‰ã§ã™ã€‚ãã‚Œã§ã¯ã€study_hoursã®[5, 8, 6, 9, 4]ã®æ•°å­—ã‚’ã€ã‚ãªãŸã®å®Ÿéš›ã®å‹‰å¼·æ™‚é–“ã«åˆã‚ã›ã¦å¤‰æ›´ã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚ãŸã¨ãˆã°ã€æœˆæ›œæ—¥ã«3æ™‚é–“ã€æœ¨æ›œæ—¥ã«7æ™‚é–“å‹‰å¼·ã—ãŸå ´åˆã¯ã€study_hoursã®ä¸€ç•ªç›®ã¨å››ç•ªç›®ã®æ•°å­—ã‚’3ã¨7ã«æ›¸ãæ›ãˆã¾ã—ã‚‡ã†ã€‚

````python
#####################################
# ğŸ”§ å¿…è¦ãªãƒ„ãƒ¼ãƒ«ã‚’æº–å‚™ã—ã‚ˆã†
#####################################
import matplotlib.pyplot as plt

#####################################
# ğŸ“Š ãƒ‡ãƒ¼ã‚¿ã‚’å…¥åŠ›ã—ã‚ˆã†
#####################################
# æ›œæ—¥ï¼ˆæ¨ªè»¸ï¼‰
days = ['Mon.', 'Tue.', 'Wed.', 'Thu.', 'Fri.']

# å‹‰å¼·æ™‚é–“ï¼ˆç¸¦è»¸ï¼‰
# ğŸŒŸã“ã“ã‚’æ›¸ãæ›ãˆã¾ã—ã‚‡ã†
study_hours = [5, 8, 6, 9, 4]

#####################################
# ğŸ¨ ã‚°ãƒ©ãƒ•ã‚’ä½œæˆ
#####################################
plt.figure(figsize=(10, 6))  # ã‚°ãƒ©ãƒ•ã®ã‚µã‚¤ã‚ºã‚’è¨­å®š
plt.plot(days, study_hours, marker='o', linewidth=2, markersize=12, color='#4CAF50')

# ğŸ·ï¸ ã‚°ãƒ©ãƒ•ã«æƒ…å ±ã‚’è¿½åŠ 
plt.title('Weekly study time', fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Day', fontsize=20)
plt.ylabel('Time [h]', fontsize=20)
plt.grid(True, alpha=0.3)  # è–„ã„ã‚°ãƒªãƒƒãƒ‰ç·š

# ğŸ“Š å„ç‚¹ã«æ•°å€¤ã‚’è¡¨ç¤º
for i, hours in enumerate(study_hours):
    plt.annotate(f'{hours}h', (days[i], hours), textcoords="offset points",
                xytext=(0,20), ha='center', fontsize=20, fontweight='bold')

plt.tight_layout()  # ãƒ¬ã‚¤ã‚¢ã‚¦ãƒˆã‚’æ•´ãˆã‚‹
plt.show()

#####################################
# ğŸ¶ ãŠã¾ã‘ï½ãƒ‡ãƒ¼ã‚¿åˆ†æï½
#####################################
# sum(study_hours)ã§study_hoursã®å€¤ã‚’ã™ã¹ã¦åŠ ç®—ã™ã‚‹
print(f"ğŸ’¡ ä¸€é€±é–“ã®åˆè¨ˆå‹‰å¼·æ™‚é–“: {sum(study_hours)}æ™‚é–“")
# len(study_hours)ã§ãƒ‡ãƒ¼ã‚¿ã‚’å…¥åŠ›ã—ãŸæ•°ã‚’å–å¾—ã™ã‚‹
print(f"ğŸ“Š å¹³å‡å‹‰å¼·æ™‚é–“: {sum(study_hours)/len(study_hours):.1f}æ™‚é–“")
````

## ğŸ”„ Step2: ã‚°ãƒ©ãƒ•ã®ç¨®é¡ã‚’å¤‰ãˆã¦ã¿ã‚ˆã†

åŒã˜ãƒ‡ãƒ¼ã‚¿ã§ã‚‚ã€ã‚°ãƒ©ãƒ•ã®å½¢å¼ãŒå¤‰ã‚ã‚‹ã¨è¦‹ãˆæ–¹ãŒå¤‰ã‚ã‚Šã¾ã™ã€‚ãŸã¨ãˆã°ã€æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ã¯å‹‰å¼·æ™‚é–“ã®æ¨ç§»ãŒä¸€ç›®ç­ç„¶ã§ã™ã€‚ãã‚Œã§ã¯ã€å‹‰å¼·æ™‚é–“ã®é•·ã•ã‚’ã‚‚ã£ã¨ã‚ã‹ã‚Šã‚„ã™ãã—ãŸã„ã¨ã—ã¾ã™ã€‚ã“ã®ã¨ãæœ‰åŠ¹ãªã‚°ãƒ©ãƒ•ã¨ã—ã¦ã€**æ£’ã‚°ãƒ©ãƒ•**ãŒæŒ™ã’ã‚‰ã‚Œã¾ã™ã€‚ã“ã¡ã‚‰ã¯æ£’ã‚°ãƒ©ãƒ•ã‚’ä½œæˆã™ã‚‹ã‚³ãƒ¼ãƒ‰ã§ã™ã€‚

````python
##############
# ãŠã¾ã˜ãªã„ #
##############

# pythonã§ã‚°ãƒ©ãƒ•ã‚’æ›¸ããŸã‚ã®é“å…·ã‚’èª­ã¿è¾¼ã‚€
import matplotlib.pyplot as plt


##############
# ãƒ‡ãƒ¼ã‚¿å…¥åŠ› #
##############

# æ›œæ—¥ï¼ˆxè»¸ï¼‰ã‚’å…¥åŠ›ã™ã‚‹
x = ['Mon.', 'Tue.', 'Wed.', 'Thu.', 'Fri.']
# å‹‰å¼·æ™‚é–“ï¼ˆyè»¸ï¼‰ã‚’å…¥åŠ›ã™ã‚‹
y = [5, 8, 6, 9, 4]


##############
# ã‚°ãƒ©ãƒ•è¨­å®š #
##############

# æ£’ã‚°ãƒ©ãƒ•ã‚’ä½œæˆã™ã‚‹ï¼ˆ"plt.plot"ã‹ã‚‰"plt.bar"ã«å¤‰ã‚ã£ãŸï¼‰
plt.bar(x, y)


##############
# ã‚°ãƒ©ãƒ•æç”» #
##############

# ã‚°ãƒ©ãƒ•ã®åå‰ã‚’ã¤ã‘ã‚‹
plt.title('Weekly study time')
# xè»¸ã®åå‰ã‚’ã¤ã‘ã‚‹
plt.xlabel('Day')
# yè»¸ã®åå‰ã‚’ã¤ã‘ã‚‹
plt.ylabel('Time [h]')
# ã‚°ãƒªãƒƒãƒ‰ç·šï¼ˆç½«ç·šï¼‰ã‚’å¼•ã
plt.grid(True)
# ã‚°ãƒ©ãƒ•ã‚’è¡¨ç¤ºã™ã‚‹
plt.show()
````

å…ˆã»ã©ã®ã‚³ãƒ¼ãƒ‰ã®```plt.plot(x, y)```ã‚’```plt.bar(x, y)```ã«å¤‰ãˆãŸã ã‘ã§ã€æŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ã‚’æ£’ã‚°ãƒ©ãƒ•ã«å¤‰æ›´ã§ãã¾ã—ãŸã€‚ã“ã®ã‚ˆã†ã«pythonã§ã¯ã€åŒã˜ãƒ‡ãƒ¼ã‚¿ã§ã‚‚æç”»æ–¹æ³•ã®éƒ¨åˆ†ã ã‘ã‚’å¤‰æ›´ã™ã‚‹ã ã‘ã§ç°¡å˜ã«ã‚°ãƒ©ãƒ•ã‚’åˆ‡ã‚Šæ›¿ãˆã‚‰ã‚Œã¾ã™ã€‚å®Ÿéš›ã«ã€æ£’ã‚°ãƒ©ãƒ•ã‚’è¦‹ã‚„ã™ãã—ãŸã‚³ãƒ¼ãƒ‰ã¯ã“ã¡ã‚‰ã§ã™ã€‚

````python
#####################################
# ğŸ”§ å¿…è¦ãªãƒ„ãƒ¼ãƒ«ã‚’æº–å‚™ã—ã‚ˆã†
#####################################
````
````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#####################################
# ğŸ“Š ä¸–ç•Œã®çµ±è¨ˆãƒ‡ãƒ¼ã‚¿ã‚’èª­ã¿è¾¼ã‚‚ã†
#####################################
# å¤–éƒ¨ã‚µãƒ¼ãƒã‹ã‚‰ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')

#####################################
# ğŸ¨ ã‚°ãƒ©ãƒ•ã‚’ä½œæˆ
#####################################
# ğŸ“Š 2007å¹´ã®ãƒ‡ãƒ¼ã‚¿ã§æ•£å¸ƒå›³ä½œæˆ
df_scatter = df[df['year'] == 2007]

# å¤§é™¸åˆ¥ã®è‰²åˆ†ã‘
continent_colors = {
    'Asia': '#e74c3c',
    'Europe': '#3498db',
    'Africa': '#f39c12',
    'Americas': '#2ecc71',
    'Oceania': '#9b59b6'
}

# ã‚°ãƒ©ãƒ•ã®å¤§ãã•
plt.figure(figsize=(14, 10))

# å¤§é™¸åˆ¥ã«æ•£å¸ƒå›³ã‚’ãƒ—ãƒ­ãƒƒãƒˆ
for continent in df_scatter['continent'].unique():
    continent_data = df_scatter[df_scatter['continent'] == continent]
    plt.scatter(continent_data['gdpPercap'], continent_data['lifeExp'],
               s=2*continent_data['pop']/1e6,  # äººå£ã‚µã‚¤ã‚ºã§å††ã®å¤§ãã•ã‚’æ±ºå®š
               c=continent_colors[continent],
               alpha=0.6, label=continent, edgecolors='black', linewidth=0.5)

# ğŸ·ï¸ ã‚°ãƒ©ãƒ•ã®è£…é£¾
plt.title('GDP Per Capita vs. Life Expectancy in 2007',
          fontsize=20, fontweight='bold', pad=20)
plt.xlabel('GDP Per Capita [USD]', fontsize=20)
plt.ylabel('Life Expectancy [Years]', fontsize=20)
plt.legend(title='Continent', fontsize=20)
plt.grid(True, alpha=0.3)

# ğŸ“ èˆˆå‘³æ·±ã„å›½ã‚’ãƒã‚¤ãƒ©ã‚¤ãƒˆ
highlight_countries = ['Japan', 'United States', 'Norway', 'Chad', 'Afghanistan']
for country in highlight_countries:
    if country in df_scatter['country'].values:
        country_data = df_scatter[df_scatter['country'] == country].iloc[0]
        plt.annotate(country,
                     (country_data['gdpPercap'], country_data['lifeExp']),
                     xytext=(5, 5), textcoords='offset points',
                     fontsize=15, fontweight='bold',
                     bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.8))

plt.tight_layout()
plt.show()


#####################################
# ğŸ¶ ãŠã¾ã‘ï½ãƒ‡ãƒ¼ã‚¿åˆ†æï½
#####################################
# ğŸ“Š ç›¸é–¢é–¢ä¿‚ã®åˆ†æ
correlation = df_scatter['gdpPercap'].corr(df_scatter['lifeExp'])
print(f"ğŸ“ˆ GDPã¨å¹³å‡å¯¿å‘½ã®ç›¸é–¢ä¿‚æ•°: {correlation:.3f}")
# ğŸ’¡ ç›¸é–¢ä¿‚æ•°ãŒ1ã«è¿‘ã„ã»ã©å¼·ã„æ­£ã®ç›¸é–¢ãŒã‚ã‚‹ã“ã¨ã‚’æ„å‘³ã™ã‚‹ï¼ˆxè»¸ã®å€¤ãŒå¤§ãããªã‚‹ã»ã©yã®å€¤ã‚‚å¤§ãããªã‚‹ï¼‰
````

````python
#####################################
# ğŸ”§ å¿…è¦ãªãƒ„ãƒ¼ãƒ«ã‚’æº–å‚™ã—ã‚ˆã†
#####################################
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import imageio.v2 as imageio
import os
from IPython.display import Image, display

#####################################
# ğŸ“Š ä¸–ç•Œã®çµ±è¨ˆãƒ‡ãƒ¼ã‚¿ã‚’èª­ã¿è¾¼ã‚‚ã†
#####################################
# å¤–éƒ¨ã‚µãƒ¼ãƒã‹ã‚‰ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')

#####################################
# ğŸ¨ ã‚°ãƒ©ãƒ•ã‚’ä½œæˆ
#####################################
# ğŸ“… è¡¨ç¤ºã™ã‚‹å¹´ã‚’é¸æŠ
years = sorted(df['year'].unique()) # ğŸŒŸ å…¨ã¦ã®å¹´ã‚’è¡¨ç¤º

# å¤§é™¸åˆ¥ã®è‰²åˆ†ã‘
continent_colors = {
'Asia':'red',
'Europe':'green',
'Africa':'blue',
'Americas':'yellow',
'Oceania':'purple'
}

# ã‚°ãƒ©ãƒ•ã®å¤§ãã•
images = []
filename_pattern = 'temp_plot_{}.png'

# å¹´ä»£åˆ¥ã«ã‚°ãƒ©ãƒ•ã‚’ä½œæˆ
for i, year in enumerate(years):
    # ãã®å¹´ã®ãƒ‡ãƒ¼ã‚¿ã‚’æŠ½å‡º
    df_year = df[(df['year'] == year)].copy() # è­¦å‘Šå›é¿ã®ãŸã‚copy()

    # æç”»è¨­å®š
    fig, ax = plt.subplots(figsize=(10, 6))

    # å¤§é™¸åˆ¥ã«ãƒ—ãƒ­ãƒƒãƒˆ
    for continent in df_year['continent'].unique():
        continent_data = df_year[(df_year['continent'] == continent)]
        ax.scatter(continent_data['gdpPercap'], continent_data['lifeExp'],
                  s=continent_data['pop']/1e6,
                  c=continent_colors.get(continent, 'gray'),
                  alpha=0.6, label=continent if year == years[-1] else "",
                  edgecolors='black', linewidth=0.5)

    # å„ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆã®è£…é£¾
    ax.set_title(f'{year}', fontsize=14, fontweight='bold')
    ax.set_xlabel('GDP Per Capita [USD]', fontsize=10)
    ax.set_ylabel('Life Expectancy [Years]', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, 50000)
    ax.set_ylim(20, 90) # yè»¸ã®ç¯„å›²ã‚’èª¿æ•´

    plt.tight_layout()

    # ç”»åƒãƒ•ã‚¡ã‚¤ãƒ«åã‚’ç”Ÿæˆã—ã¦ä¿å­˜
    filename = filename_pattern.format(year)
    plt.savefig(filename)
    images.append(imageio.imread(filename))
    plt.close()

#####################################
# ğŸ“¹ GIFã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã‚’ä½œæˆ
#####################################
# GIFã®ä½œæˆ
imageio.mimsave('gdp_vs_lifeExp.gif', images, duration=1000, loop=0)

# GIFã‚’è¡¨ç¤º
display(Image(filename='gdp_vs_lifeExp.gif'))

# ä¸€æ™‚ãƒ•ã‚¡ã‚¤ãƒ«ã‚’å‰Šé™¤
for year in years:
    filename = filename_pattern.format(year)
    os.remove(filename)
````

````python
#####################################
# ğŸ”§ å¿…è¦ãªãƒ„ãƒ¼ãƒ«ã‚’æº–å‚™ã—ã‚ˆã†
#####################################
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#####################################
# ğŸ“Š ä¸–ç•Œã®çµ±è¨ˆãƒ‡ãƒ¼ã‚¿ã‚’èª­ã¿è¾¼ã‚‚ã†
#####################################
# å¤–éƒ¨ã‚µãƒ¼ãƒã‹ã‚‰ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')

# ğŸŒŸ ã“ã“ã‚’å¤‰æ›´ã—ã¦ã€ãŠæ°—ã«å…¥ã‚Šã®å›½ã‚’è¨­å®šã—ã‚ˆã†ï¼
favorite_countries = ['Japan', 'Singapore', 'United States']
target_year = 2007

# ğŸ“Š ãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰ä½œæˆ
fig = plt.figure(figsize=(12, 20))

# ãƒ¬ã‚¤ã‚¢ã‚¦ãƒˆ: 2x2 ã®ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ
ax1 = plt.subplot(3, 2, 1)  # äººå£æ¯”è¼ƒ
ax2 = plt.subplot(3, 2, 2)  # GDPæ¯”è¼ƒ
ax3 = plt.subplot(3, 2, 3)  # å¹³å‡å¯¿å‘½æ¯”è¼ƒ
ax4 = plt.subplot(3, 2, 4)  # äººå£æ¨ç§»
ax5 = plt.subplot(3, 2, 5)  # å¹³å‡å¯¿å‘½æ¯”è¼ƒ
ax6 = plt.subplot(3, 2, 6)  # äººå£æ¨ç§»

colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12']

# ğŸ“Š ã‚°ãƒ©ãƒ•1: äººå£æ¯”è¼ƒï¼ˆæ£’ã‚°ãƒ©ãƒ•ï¼‰
pop_data = []
for country in favorite_countries:
    pop = df[(df['country'] == country) & (df['year'] == target_year)]['pop'].iloc[0]
    pop_data.append(pop / 1e6)

bars1 = ax1.bar(favorite_countries, pop_data, color=colors)
ax1.set_title(f'Population ({target_year})', fontweight='bold')
ax1.set_ylabel('Population [million]')
ax1.tick_params(axis='x', rotation=45)
for i, v in enumerate(pop_data):
    ax1.text(i, v + max(pop_data)*0.01, f'{v:.0f} M', ha='center', fontweight='bold')

# ğŸ“Š ã‚°ãƒ©ãƒ•2: GDPæ¯”è¼ƒï¼ˆæ£’ã‚°ãƒ©ãƒ•ï¼‰
gdp_data = []
for country in favorite_countries:
    gdp = df[(df['country'] == country) & (df['year'] == target_year)]['gdpPercap'].iloc[0]
    gdp_data.append(gdp)

bars2 = ax2.bar(favorite_countries, gdp_data, color=colors)
ax2.set_title(f'GDP Per Capita ({target_year})', fontweight='bold')
ax2.set_ylabel('GDP Per Capita [USD]')
ax2.tick_params(axis='x', rotation=45)
for i, v in enumerate(gdp_data):
    ax2.text(i, v + max(gdp_data)*0.01, f'${v:,.0f}', ha='center', fontweight='bold')

# ğŸ“Š ã‚°ãƒ©ãƒ•3: å¹³å‡å¯¿å‘½æ¯”è¼ƒï¼ˆæ£’ã‚°ãƒ©ãƒ•ï¼‰
life_data = []
for country in favorite_countries:
    life = df[(df['country'] == country) & (df['year'] == target_year)]['lifeExp'].iloc[0]
    life_data.append(life)

bars3 = ax3.bar(favorite_countries, life_data, color=colors)
ax3.set_title(f'Life Expectancy ({target_year})', fontweight='bold')
ax3.set_ylabel('Life Expectancy [years]')
ax3.tick_params(axis='x', rotation=45)
for i, v in enumerate(life_data):
    ax3.text(i, v + max(life_data)*0.01, f'{v:.1f} years', ha='center', fontweight='bold')

# ğŸ“Š ã‚°ãƒ©ãƒ•4: äººå£æ¨ç§»ï¼ˆæŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ï¼‰
for i, country in enumerate(favorite_countries):
    country_data = df[df['country'] == country]
    ax4.plot(country_data['year'], country_data['pop'] / 1e6,
             marker='o', linewidth=2, markersize=6,
             color=colors[i], label=country)

ax4.set_title('Population (1952-2007)', fontweight='bold')
ax4.set_xlabel('Year')
ax4.set_ylabel('Population [million]')
ax4.legend()
ax4.grid(True, alpha=0.3)

# ğŸ“Š ã‚°ãƒ©ãƒ•5: äººå£æ¨ç§»ï¼ˆæŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ï¼‰
for i, country in enumerate(favorite_countries):
    country_data = df[df['country'] == country]
    ax5.plot(country_data['year'], country_data['gdpPercap'],
             marker='o', linewidth=2, markersize=6,
             color=colors[i], label=country)

ax5.set_title('GDP Per Capita (1952-2007)', fontweight='bold')
ax5.set_xlabel('Year')
ax5.set_ylabel('GDP Per Capita [USD]')
ax5.legend()
ax5.grid(True, alpha=0.3)

# ğŸ“Š ã‚°ãƒ©ãƒ•6: äººå£æ¨ç§»ï¼ˆæŠ˜ã‚Œç·šã‚°ãƒ©ãƒ•ï¼‰
for i, country in enumerate(favorite_countries):
    country_data = df[df['country'] == country]
    ax6.plot(country_data['year'], country_data['lifeExp'],
             marker='o', linewidth=2, markersize=6,
             color=colors[i], label=country)

ax6.set_title('Life Expectancy (1952-2007)', fontweight='bold')
ax6.set_xlabel('Year')
ax6.set_ylabel('Life Expectancy [years]')
ax6.legend()
ax6.grid(True, alpha=0.3)

# å…¨ä½“ã®ã‚¿ã‚¤ãƒˆãƒ«
fig.suptitle(f'Dashboard: {" vs ".join(favorite_countries)}',
             fontsize=20, fontweight='bold', y=1.0)

plt.tight_layout()
plt.show()


````
